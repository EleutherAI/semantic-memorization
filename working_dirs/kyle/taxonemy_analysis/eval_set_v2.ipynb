```python
import pandas as pd
import numpy as np
from datasets import load_dataset
from transformers import AutoTokenizer
```

```python
eval_set = pd.read_csv("Pythia_70m_Deduped_Low_Perplexity_Labeling_Formatted.csv")
eval_set
```

```python
pythia_70m_memories = load_dataset("EleutherAI/pythia-memorized-evals", split="deduped.70m").to_pandas()
pythia_70m_memories
```

```python
# join both frames on their column named index
memories_eval_set = eval_set.join(pythia_70m_memories.set_index('index'), on='index', how="inner")
memories_eval_set
```

```python
non_memories = load_dataset("EleutherAI/pile-deduped-pythia-random-sampled")["train"].to_pandas()
non_memories
```

```python
non_memories_eval_set = eval_set.join(non_memories.set_index('index'), on='index', how="inner")
non_memories_eval_set
```

```python
tokenizer = AutoTokenizer.from_pretrained("EleutherAI/pythia-70m-deduped")
```

```python
target_columns = ["index", "memorized", "perplexity", "is_code", "tokens"]
combined_joined_eval_set = pd.concat([memories_eval_set, non_memories_eval_set])[target_columns]
combined_joined_eval_set["prompt"] = combined_joined_eval_set["tokens"].apply(lambda x: tokenizer.decode(x[:32]))
combined_joined_eval_set["sequence"] = combined_joined_eval_set["tokens"].apply(lambda x: tokenizer.decode(x[:64]))
combined_joined_eval_set = combined_joined_eval_set.drop_duplicates(subset=["index"]).drop(columns=["tokens"]).sample(frac=1)

annotater_columns = [
    "Code",
    "Incremental",
    "Repetitive",
    "Highly Duplicated",
    "Templating",
    "Natural Language",
    "Random",
    "Other",
    "Notes"
]

combined_joined_eval_set.to_csv("combined_formatted_eval_set.csv", index=False)
for col in annotater_columns:
    combined_joined_eval_set[col] = ""
combined_joined_eval_set
```

```python
combined_strat_downsample = pd.concat(
    [combined_joined_eval_set[combined_joined_eval_set["memorized"]].sample(50),
     combined_joined_eval_set[~combined_joined_eval_set["memorized"]].sample(50)]).sample(frac=1)

combined_strat_downsample.to_csv("combined_strat_downsample.csv", index=False)
combined_strat_downsample
```
